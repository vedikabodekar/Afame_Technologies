```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
data['Year'] = data['Order Date'].dt.year
data['Year']
```

```python
data_overtime = data.groupby(['Year', 'Category'])['Sales'].sum().round(2)
data_overtime
```

```python
data_overtime.unstack().plot(kind='line', figsize=(10, 8))
plt.title('Sales Performance of Different Product Categories Over Time')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.legend(title='Category')
plt.grid(True)
plt.show()
```

************Which customer segments contribute the most to overall sales revenue?************

```python
data.columns
```

```python
highest_segment = data.groupby(['Segment'])['Sales'].sum().reset_index().round(2)
highest_segment
```

```python
plt.bar(highest_segment['Segment'],highest_segment['Sales'], color = "green")
plt.xlabel("Segment")
plt.ylabel("Totla Sales")
plt.title('Customer segments contribute the most to overall sales revenue')
plt.show()
```

**Are there any noticeable patterns in sales based on geographic regions or market segments?**

```python
sales_by_segments = data.groupby(['Segment'])['Sales'].sum().round()
sales_by_segments

sales_by_segments.plot(kind='bar', color= 'pink')
plt.show()
```
